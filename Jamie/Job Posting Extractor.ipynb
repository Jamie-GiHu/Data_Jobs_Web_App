```python
import requests
import pandas as pd
import re
import time
from config import api_key,api_id
```

```python
def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
def getJsonResponse(country,State,keyword,page):
    url = f'https://api.adzuna.com/v1/api/jobs/au/search/{page}?app_id={api_id}&app_key={api_key}&title_only={keyword}&location0={country}&location1={State}'
    headers = {'Content-type': 'application/json'}
    time.sleep(0.25)
    resp = requests.get(url, headers=headers).json()  
    return resp
```
